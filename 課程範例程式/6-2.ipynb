```python
import re
```

```python
pattern = re.compile(r"a?\d+\+?")
str1 = "123 a4567 4567+"
result = pattern.search(str1)
result.group()
```

```python
result2 = re.search(pattern, str1)
result2.group()
```

```python
pattern = re.compile(r"(\d+) (\d+)")
str1 = "123 4567 789"
result = pattern.search(str1)
result.group()
```

```python
result.group(0)
```

```python
result.group(1)
```

```python
result.group(2)
```

```python
pattern = re.compile(r"(\d+) (\d+)")
str1 = "123 4567 789"
re.match(pattern, str1).group()
```

```python
pattern = re.compile(r"(\d{4}) (\d+)")
str1 = "123 4567 789"
result = re.match(pattern, str1)
print(result)
print()
```

```python
pattern_g = re.compile(r"\d+ \d+")
str1 = "123 4567 789 45"
result = re.findall(pattern, str1)
print(result[0][0])
print(result)
```

```python
pattern = re.compile(r"(\d+?) (\d+?)")
str1 = "123 4567 789 45"
result = re.findall(pattern, str1)
print(result[0][0])
print(result)
```
