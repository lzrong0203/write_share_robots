```python
from matplotlib import pyplot as plt
import numpy as np
import numpy.random as nr
```

```python
nr.seed(0)
dice1 = np.ceil(nr.rand(100) * 6).astype(int)
dice1
```

```python
nr.seed(0)
dice2 = nr.randint(1, 7, 100)
dice2
```

```python
plt.hist(dice1, bins=6)
plt.show()
```

```python
plt.hist(dice2, bins=6)
plt.show()
```

```python
nr.seed(0)
dice3 = np.ceil(nr.rand(10000) * 6).astype(int)
```

```python
plt.hist(dice3, bins=6)
plt.show()
```

```python
plt.hist(dice3, bins=6, edgecolor="black")
plt.title("DICE ROLLING")
plt.ylabel("Times")
plt.xlabel("dice number")
plt.grid()
plt.show()
```

```python
nr.seed(0)
normal = nr.randn(100000)
mean = np.mean(normal)
std = np.std(normal)
# plt.vlines(mean, np.min(normal), np.max(normal))
plt.axvline(mean, color="r")
plt.axvline(mean - 2 * std, color="black")
plt.axvline(mean + 2 * std, color="black")
plt.hist(normal, histtype="stepfilled", edgecolor="b", linestyle="--", linewidth=2)
plt.show()
```

```python
import yfinance as yf

tsmc = yf.download("2330.tw", start="2023-01-01", end="2023-12-31")
tsmc['Change'] = tsmc['Close'].diff()
tsmc["K_Change"] = tsmc.Close - tsmc.Open
```

```python
tsmc_Change_mean = np.mean(tsmc.Change)
print(np.mean(tsmc.Close))
print(tsmc_Change_mean)
tsmc_Change_std = np.std(tsmc.Change)
print(tsmc_Change_std)
plt.axvline(tsmc_Change_mean, color="r")
plt.axvline(tsmc_Change_mean - 2 * tsmc_Change_std, color="black")
plt.axvline(tsmc_Change_mean + 2 * tsmc_Change_std, color="black")
plt.hist(tsmc.Change, bins=20, edgecolor="black")
plt.show()
```

```python
tsmc_KChange_mean = np.mean(tsmc.K_Change)
print(np.mean(tsmc.Close))
print(tsmc_KChange_mean)
tsmc_KChange_std = np.std(tsmc.K_Change)
print(tsmc_KChange_std)
plt.axvline(tsmc_KChange_mean, color="r")
plt.axvline(tsmc_KChange_mean - 2 * tsmc_KChange_std, color="black")
plt.axvline(tsmc_KChange_mean + 2 * tsmc_KChange_std, color="black")
plt.hist(tsmc.K_Change, edgecolor="black")
plt.show()
```
