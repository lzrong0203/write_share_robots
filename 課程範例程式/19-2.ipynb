GDP

```python
gdp_url = "https://nstatdb.dgbas.gov.tw/dgbasall/webMain.aspx?sdmx/a018101010/4...A.&startTime=2011&endTime=2023"
```

```python
import requests
import pandas as pd
```

```python
r = requests.get(gdp_url)
```

```python
r.status_code
```

```python
data = pd.read_json(r.text)
```

```python
data
```

```python
gdp = data.data.dataSets[0]["series"]["0"]["observations"]
```

```python
time_periods = data.data.structure["dimensions"]["observation"][0]["values"]
```

```python
gdp
```

```python
time_periods
```

```python
gdp_data = []
for i, p in enumerate(time_periods):
    gdp_data.append([time_periods[i]["id"], gdp[str(i)][0]])
```

```python
gdp_df = pd.DataFrame(gdp_data, columns=["year", "GDP"])
```
